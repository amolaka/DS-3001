**Q1.** This question is a case study for $k$ nearest neighbor The target variable `y` is `price` and the features are `year` and `mileage`.

1. Load the `./data/USA_cars_datasets.csv`. Keep the following variables and drop the rest: `price`, `year`, `mileage`. Are there any `NA`'s to handle? Look at the head and dimensions of the data.
2. Maxmin normalize `year` and `mileage`.
3. Split the sample into ~80% for training and ~20% for evaluation.
4. Use the $k$NN algorithm and the training data to predict `price` using `year` and `mileage` for the test set for $k=3,10,25,50,100,300$. For each value of $k$, compute the mean squared error and print a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?
5. Determine the optimal $k$ for these data.
6. Describe what happened in the plots of predicted versus actual prices as $k$ varied, taking your answer into part 6 into account. (Hint: Use the words "underfitting" and "overfitting".)

````python
# Import packages
# ! git clone https://github.com/DS3001/knn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load and look at data
cars_df = pd.read_csv('/content/knn/data/USA_cars_datasets.csv', low_memory = False)
cars_df = cars_df.loc[:, ['price', 'year', 'mileage']]
print(cars_df.shape)
cars_df.head()
````

````python
# Look for NAs
cars_df.describe()
# There are no NAs, since there are 2499 observations of each variable and 2499 rows in the dataset
````

````python
# Maxmin normalize
def maxmin(x):
  z = (x - min(x)) / (max(x) - min(x))
  return z
cars_df['year'] = maxmin(cars_df['year'])
cars_df['mileage'] = maxmin(cars_df['mileage'])
````

````python
# Split sample
from sklearn.model_selection import train_test_split
y = cars_df['price']
X = cars_df.drop('price', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
````

````python
# knn
from sklearn.neighbors import KNeighborsRegressor

for k in [3, 10, 25, 50, 100, 300]:
    model = KNeighborsRegressor(n_neighbors = k)
    fitted_model = model.fit(X_train, y_train)
    y_hat = model.predict(X_test)
    SSE = np.sum((y_test  -y_hat) ** 2 )

    plt.scatter(y_test, y_hat)
    plt.title('k: '+ str(k))
    plt.ylim(-1000, 55000)
    plt.xlim(-1000, 65000)
    plt.show()
````

````python
# optimal k
k_bar = 250
k_nums = np.arange(1, k_bar)
SSE = np.zeros(k_bar)

for k in range(k_bar):
  model = KNeighborsRegressor(n_neighbors = k + 1)
  fitted_model = model.fit(X_train, y_train)
  y_hat = fitted_model.predict(X_test)
  SSE[k] = np.sum((y_test - y_hat) ** 2)

SSE_min = np.min(SSE)
min_index = np.where(SSE == SSE_min)
k_star = k_nums[min_index]
print(k_star)

plt.plot(np.arange(0, k_bar), SSE)
plt.xlabel("k")
plt.ylabel("SSE")
plt.title("Optimal K: " + str(k_star))
plt.show()
````

Since the optimal k value we got was 77, the models where k is equal to 3, 10, and 25 are underfitting. The model where k is equal to 100 is overfitting. The models where k is equal to 50 and 100 are pretty close to the optimal model, but not completely acurate. For high k values, the scatterplot seem to bunch up together horizontally, whereas for lower k values, the scatterplots seemed to be very spread out and not as precise. As k increased, the graphs became more bunched up and outliers were less apparent.

````python
# Build a matrix
vars = ['age', 'anaemia', 'ejection_fraction', 'high_blood_pressure', 'serum_creatinine']
y = heart_df['DEATH_EVENT']
X = heart_df.loc[:, vars]
````

````python
# Maxmin normalize
def maxmin(x):
  z = (x - min(x)) / (max(x) - min(x))
  return z
X = X.apply(maxmin)
````

````python
# Split samples
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 100)
````

````python
# knn model
from sklearn.neighbors import KNeighborsRegressor
k_bar = 25
k_nums = np.arange(1, k_bar)
SSE = np.zeros(k_bar)

for k in range(k_bar):
  model = KNeighborsRegressor(n_neighbors = k+1)
  fitted_model = model.fit(X_train, y_train)
  y_hat = model.predict(X_test)
  SSE[k] = np.sum((y_test - y_hat) ** 2)

SSE_min = np.min(SSE)
min_index = np.where(SSE == SSE_min)
k_star = k_nums[min_index]
print(k_star)
print(SSE_min)

plt.plot(np.arange(0, k_bar), SSE)
plt.xlabel("k")
plt.ylabel("SSE")
plt.title('Optimal K: ' + str(k_star))
plt.show()
````

````python
# knn model for all variables
X = heart_df.drop("DEATH_EVENT", axis = 1)
X = X.apply(maxmin)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)

k_bar = 100
k_nums = np.arange(1, k_bar)
SSE = np.zeros(k_bar)

for k in range(k_bar):
  knn = KNeighborsRegressor(n_neighbors = k+1)
  fitted_knn = knn.fit(X_train, y_train)
  y_hat = knn.predict(X_test)
  SSE[k] = np.sum((y_test - y_hat) ** 2)

SSE_min = np.min(SSE)
min_index = np.where(SSE == SSE_min)
k_star = k_nums[min_index]
print(k_star)
print(SSE_min)

plt.plot(np.arange(0, k_bar), SSE)
plt.xlabel("k")
plt.ylabel("SSE")
plt.title("Optimal K: " + str(k_star))
plt.show()
````

When we have more variables, the knn model chose a higher optimal k* value than when we just had a vew variables. The SSE is slightly lower (so better) for the model with more variables, but both seem pretty similar in their ability to predict.
